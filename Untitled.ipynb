```python
# encoding the categorical columns
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
data['Card Type'] = encoder.fit_transform(data[['Card Type']])

data['Gender'].replace({'Female':1,'Male':0},inplace=True)
data['Geography'].replace({'France':0,'Spain':1,'Germany':2},inplace=True)
```

```python
plt.figure(figsize=(10,8))
sbn.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
```

```python
data_correlations=data.corr()['Exited'].drop('Exited')

data_correlations = data_correlations.sort_values(ascending=False)
plt.title("Correlation with Exited")
ax = data_correlations.plot(kind='bar',color='blue')
```

```python
complained_stayed = data[(data['Complain'] == 1) & (data['Exited'] == 0)]
complained_left = data[(data['Complain'] == 1) & (data['Exited'] == 1)]

stayed_count = complained_stayed.shape[0]
left_count = complained_left.shape[0]

plt.pie([stayed_count,left_count],labels=['Stayed','Left'])
plt.legend()

#shows that most of those who complained left
```

```python
y = data['Exited']
X= data.drop(['Exited'],axis=1)

X.shape,y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report

X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.3,random_state=42)

scaler = MinMaxScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# handling class imbalance
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=20)

X_train_smote,y_train_smote=smote.fit_resample(X_train_scaled, y_train)

X_train_smote.shape
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import tensorflow as tf
```

```python
model = Sequential()

model.add(Dense(14,input_shape=(14,),activation='relu'))

model.add(Dense(6,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train_smote,y_train_smote,epochs=20,batch_size=100,validation_data=(X_test_scaled,y_test))
```

```python
y_pred=model.predict(X_test_scaled)
y_pred[:5]
```

```python
y_pred_binary = (y_pred > 0.5).astype(int)
```

```python
accuracy = accuracy_score(y_test, y_pred_binary)
print(f'Accuracy: {accuracy:.2f}')
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_binary)

plt.figure(figsize=(8,6))
sbn.heatmap(cm,annot=True,fmt='d')
plt.xlabel("predicted")
plt.ylabel("actual")
plt.show()
```

```python
print(classification_report(y_test,y_pred_binary))
```
